```python
ts1=df_swap_curve[['Date',target]].set_index('Date').sort_index(ascending=True)
start_date ,end_date = '2016-12-29','2022-05-04'
date_mask = (ts1.index >= start_date) & (ts1.index < end_date)
dates=ts1.index[date_mask] 
ts1=ts1.loc[dates]
```

```python
# create a differenced series
def difference(dataset, interval=1):
	diff = []
	for i in range(interval, len(dataset)):
		value = dataset[i] - dataset[i - interval]
		diff.append(value)
	return pd.Series(diff)
```

```python
# transform data to be stationary
raw_values = ts1[target]
diff_values = difference(raw_values, 1)
ts1_diff=pd.DataFrame()
ts1_diff[target]=diff_values
ts1_diff.index=ts1[1:].index
```

```python
def create_dataset(X, y, n_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - n_steps):
        v = X.iloc[i:(i + n_steps)].values
        Xs.append(v)
        ys.append(y.iloc[i + n_steps])
    return np.array(Xs), np.array(ys)
```

```python
n_steps = 5
n_features = 1
```

```python
train_size = int(len(ts1_diff) * 0.85)
test_size = len(ts1_diff) - train_size
train, test = ts1_diff.iloc[0:train_size], ts1_diff.iloc[train_size:len(ts1_diff)]
print(len(train), len(test))
```

```python
def scale(x_train,y_train,x_test,y_test):
	# fit scaler
	scaler = MinMaxScaler(feature_range=(0, 1))
	scaler = scaler.fit(train.reshape(-1, train.shape[-1]))
	# transform train
	train_scaled = scaler.transform(train.reshape(-1, train.shape[-1])).reshape(train.shape)
	# transform test
	test_scaled =  scaler.transform(test.reshape(-1, test.shape[-1])).reshape(test.shape)
	return scaler, train_scaled, test_scaled
 
# inverse scaling for a forecasted value
def invert_scale(scaler, X, yhat):
	new_row = [x for x in X] + [yhat]
	array = numpy.array(new_row)
	array = array.reshape(1, len(array))
	inverted = scaler.inverse_transform(array)
	return inverted[0, -1]
```

```python
# reshape to [samples, n_steps, n_features]
X_train, y_train = create_dataset(train, train[target], n_steps)
X_test, y_test = create_dataset(test, test[target], n_steps)

print(f'X_train shape:{X_train.shape}  y_train shape:{y_train.shape}')
print(f'X_test  shape:{X_test.shape}   y_test  shape:{y_test.shape}')
```

```python
# scaler = MinMaxScaler(feature_range=(0, 1))
# X_train_scaled = scaler.fit_transform(X_train.reshape(-1, X_train.shape[-1])).reshape(X_train.shape)
# X_test_scaled = scaler.transform(X_test.reshape(-1, X_test.shape[-1])).reshape(X_test.shape)
```

```python
# define model
model = Sequential()
model.add(LSTM(8, activation='sigmoid',return_sequences=True, input_shape=(n_steps, n_features)))
model.add(LSTM(8, activation='sigmoid'))
model.add(Dense(1, activation='linear'))
opt=Adam(learning_rate=0.0001)
model.compile(optimizer=opt, loss='mse', metrics='mae')
```

```python
# early stopping
early_stopping_monitor = EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=5,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=True
)
```

```python
# fit model
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test), 
                    epochs=200, 
                    batch_size=3,
                    shuffle=False,
                    verbose=1,
                    callbacks=early_stopping_monitor
                   
)
# callbacks=early_stopping_monitor
# validation_data=(X_test, y_test)
# validation_split=0.1, 
```

```python
# evaluate the model
train_mae = model.evaluate(X_train, y_train, verbose=0)[1]
test_mae = model.evaluate(X_test, y_test, verbose=0)[1]
print(f'Train MAE: {train_mae:.3f}, Test MAE: {test_mae:.3f}')
```

```python
df=pd.DataFrame()
df['train'] = history.history['mae']
df['test'] = history.history['val_mae']
df['epoch']=[i for i in range(1,len(df['train'])+1)]

fig = px.line(df,x='epoch',y=['train','test'],labels={'value':'mae'}, title='Train Test MAE')
fig.update_yaxes(nticks=10)
fig.update_layout(width=600,height=400)
fig.show()
```

```python
y_pred = model.predict(X_test)  
df=pd.DataFrame(y_pred,columns = ['pred'])
df['actual']=y_test
df.index=test.index[n_steps:]

fig = px.line(df,x=df.index,y=['actual','pred'],title='actual vs pred')
fig.update_layout(legend_title=target, width=800,height=600)
# legend=dict(yanchor="top",y=0.98,xanchor="left",x=0.02)
fig.show()
```

```python
# pred on last 5 days for tomorow pred
model.predict(test.tail(n_steps).values.reshape(1,n_steps,1))
```
