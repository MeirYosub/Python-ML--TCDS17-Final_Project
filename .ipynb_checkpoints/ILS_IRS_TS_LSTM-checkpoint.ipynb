```python
df_ILSIRS_curve = pd.read_csv('./data/ILSIRS_AM3T_data.csv',parse_dates=['Date'])
```

```python
df_ILSIRS_curve['Date'].duplicated().sum()
```

```python
target='ILS10Y'
cols=['Date',target]
ts1=df_ILSIRS_curve[cols]
ts1=ts1.set_index('Date').sort_index(ascending=True)
start_date ,end_date = '2011-12-19','2022-05-11'
ts1=ts1[start_date:end_date]
ts1
```

```python
ts1.isnull().sum()
```

```python
n_lag=1
lag_col='lag_'+str(n_lag)
ts1[lag_col]=ts1[target].shift(n_lag)
```

```python
n_lag=2
sma2='SMA'+str(n_lag)
ts1[sma2] = ts1[target].rolling(n_lag,closed='left').mean()
```

```python
n_lag=5
sma5='SMA'+str(n_lag)
ts1[sma5] = ts1[target].rolling(n_lag,closed='left').mean()
```

```python
n_lag=10
sma10='SMA'+str(n_lag)
ts1[sma10] = ts1[target].rolling(n_lag,closed='left').mean()
```

```python
train,test = ts_train_test_split(ts1,0.9)
print(len(train), len(test))
test
```

```python
# baseline n_lag- pred on prev n_lag days
lag1_mae=mean_absolute_error(test[target], test['lag_1'])
print(f'MAE baseline - pred on lag_1 : {lag1_mae:.3f}')
```

```python
# baseline MAE- simple moving avarage w/ lag_2
sma2_mae=mean_absolute_error(test[target], test['SMA2'])
print(f'MAE baseline - pred on SMA{n_lag} : {sma2_mae:.3f}')
```

```python
# baseline MAE- simple moving avarage w/ lag_5
sma5_mae=mean_absolute_error(test[target], test['SMA5'])
print(f'MAE baseline for SMA{n_lag} : {sma5_mae:.3f}')
```

```python
# baseline MAE- simple moving avarage w/ lag_10
sma10_mae=mean_absolute_error(test[target], test['SMA10'])
print(f'MAE baseline for SMA{n_lag} : {sma10_mae:.3f}')
```

```python
fig = px.line(test,x=test.index,y=[target,sma2,sma5,sma10],
              title=f'ILS 10Y actual vs SMA pred- baseline')
fig.update_layout(legend_title=target, width=800,height=600)
fig.show()
```

```python
mtm_lag1=trade_on_pred(test[target],test['lag_1'],lag1_mae)
mtm_ma2=trade_on_pred(test[target],test[sma2],sma2_mae)
mtm_ma5=trade_on_pred(test[target],test[sma5],sma5_mae)
mtm_ma10=trade_on_pred(test[target],test[sma10],sma10_mae)

print('Trade P&L results on test data w/ SMA pred:')
print(f'Baseline 0 - lag_1 pred : {mtm_lag1:.2f}')
print(f'Baseline 1 - {sma2}  pred : {mtm_ma2:.2f}')
print(f'Baseline 2 - {sma5}  pred : {mtm_ma5:.2f}')
print(f'Baseline 3 - {sma10} pred : {mtm_ma10:.2f}')

```

```python
ts2=df_ILSIRS_curve[cols]
ts2=ts2.set_index('Date').sort_index(ascending=True)
start_date ,end_date = '2012-01-01','2022-05-11'
ts2=ts2[start_date:end_date]
ts2
```

```python
train,test = ts_train_test_split(ts2,0.9)
print(len(train), len(test))
```

```python
n_lag=10
n_features = 1
```

```python
# reshape to [samples, n_steps, n_features]
X_train, y_train = create_dataset(train, train[target], n_lag)
X_test, y_test = create_dataset(test, test[target], n_lag)

print(f'X_train shape:{X_train.shape}  y_train shape:{y_train.shape}')
print(f'X_test  shape:{X_test.shape}   y_test  shape:{y_test.shape}')
```

```python
# early stopping
early_stopping_monitor = EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=5,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=True
)
```

```python
# define and fir the model
model, history=fit_LSTM_type1(
    X_train,y_train,X_test,y_test,n_lag,n_features,
    l1_neurons=8,
    batch_size=3,
    n_epochs=100,
    learning_rate=0.0001,
    verbose=0,
    callbacks=early_stopping_monitor
)
```

```python
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_dtype=False, show_shapes=True, show_layer_names=True)
```

```python
# plot mlp learnig curve
title='MLP learing curve - Train vs Test MAE'
mlp_learnig_curve_plot(history,title).show()
```

```python
# evaluate the model
train_mae, test_mae = evaluate_model(model, X_train, y_train, X_test, y_test, verbose=0)
print(f'Train MAE: {train_mae:.3f}, Test MAE: {test_mae:.3f}')
```

```python
# test set - actual vs predicted plot
y_pred = model.predict(X_test)
title=f'{target} - actual vs predicted'
test_predicted_plot(y_pred, y_test, test.index[n_lag:], title,target).show()
```

```python
df_SPXTR = pd.read_csv('./data/SPXTR.csv',parse_dates=['Date']).dropna(how='any').set_index('Date').sort_index(ascending=True)
df_SPXTR
```

```python
df_SPXTR['SP_yield']=df_SPXTR.SPXTR.pct_change()
df_SPXTR
```

```python
ts4=ts3.merge(df_SPXTR['SP_yield'],how='left', on='Date').dropna(how='any')
ts4
```

```python
ts4.isnull().sum()
```

```python
n_lag = 10
n_features = 3
```

```python
train,test = ts_train_test_split(ts4,0.9)
print(len(train), len(test))
```

```python
train=train.copy()
test=test.copy()
# fit scaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaler = scaler.fit(train['SP_yield'].values.reshape(-1,1))
# transform train
train['SP_yield_sc']=pd.Series(scaler.transform(train['SP_yield'].values.reshape(-1,1)).reshape(-1), index=train.index)
train=train.drop('SP_yield',axis=1)

# transform test
test['SP_yield_sc']=pd.Series(scaler.transform(test['SP_yield'].values.reshape(-1,1)).reshape(-1), index=test.index)
test=test.drop('SP_yield',axis=1)
```

```python
# reshape to [samples, time_steps, n_features]
X_train, y_train = create_dataset(train, train[target], n_lag)
X_test, y_test = create_dataset(test, test[target], n_lag)

print(f'X_train shape:{X_train.shape}  y_train shape:{y_train.shape}')
print(f'X_test  shape:{X_test.shape}   y_test  shape:{y_test.shape}')
```

```python
# define and fit the model 
model, history=fit_LSTM_type1(
    X_train,y_train,X_test,y_test,n_lag,n_features,
    l1_neurons=16,
    batch_size=3,
    n_epochs=100,
    learning_rate=0.0001,
    verbose=0,
    callbacks=early_stopping_monitor
)
```

```python
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_dtype=False, show_shapes=True, show_layer_names=True)
```

```python
# plot mlp learnig curve
title='MLP learing curve - Train vs Test MAE'
mlp_learnig_curve_plot(history,title).show()
```

```python
# evaluate the model
train_mae = model.evaluate(X_train, y_train, verbose=0)[1]
test_mae = model.evaluate(X_test, y_test, verbose=0)[1]
print(f'Train MAE: {train_mae:.3f}, Test MAE: {test_mae:.3f}')
```

```python
# test set - actual vs predicted plot
y_pred = model.predict(X_test)
title=f'{target} - actual vs predicted w/ USD10Y and S&P500 as feauters'
test_predicted_plot(y_pred, y_test, test.index[n_lag:], title,target).show()
```

```python
mtm_nn_3=trade_on_pred(y_test,y_pred,test_mae)
print(f'Trade P&L : {mtm_nn_3:.2f}')
```
